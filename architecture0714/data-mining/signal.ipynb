```python
import matplotlib.pyplot as plt
import numpy 


pi=np.pi


f1 = [ 0, 0, 0.1, 0.3, 13.3, 33.3 ]
f2 = [ 0.5, 6.5, 16.6, 17.8, 29.8, 49.8 ]

pi=math.pi

signalTime = numpy.arange(0, 6, .01)
plt.figure(figsize=(16,16))

for i in range(len(f1)):
    
    plt.subplot(6,1,i+1)
    s1_a  = numpy.sin(2*pi*signalTime*f1[i])
    s2_a  = numpy.sin(2*pi*signalTime*f2[i]) 
    merge_sig_amp = s1_a+s2_a
    
    plt.plot(signalTime, merge_sig_amp, color ='green') 
    plt.xlabel('Time') 
    plt.ylabel('Amplitude') 
```
